```python
# IMPORTA AS BIBLIOTECAS

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados da cafeteria

df = pd.read_csv('cafeteria.csv')
print(df)
```

### Temas do Seaborn

- whitegrid = grade branca (fundo branco com grad)
- darkgrid = fundo cinza com grade
- white = fundo branco sem grade


para usar tema: `sns.set_style('nome do tema')`

### Scatter

- Scatter = gráfico de dispersão

```python
# usando um tema
sns.set_style('whitegrid')


# ===============================
# GRÁFICO DE DISPERSÃO
# ===============================

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='preco', y='avaliacao')
plt.title('Preço x Avaliação',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Preço', fontsize=12)
plt.ylabel('Avaliação', fontsize=12)
plt.show()
```

```python
# ===============================
# GRÁFICO DE DISPERSÃO COM CORES
# ===============================


# hue = matiz

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='preco', y='avaliacao', hue='periodo')
plt.title('Preço x Avaliação',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Preço', fontsize=12)
plt.ylabel('Avaliação', fontsize=12)
plt.show()
```

```python
# ===============================
# GRÁFICO DE DISPERSÃO COM TAMANHOS DIFERENTES
# ===============================

# SIZE = TAMANHO
# Bolinha maior = mais vendas.
# Bolinha menor = menos vendas.

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='preco', y='avaliacao', hue='periodo', size='vendas', sizes=(100, 400))
plt.title('Preço x Avaliação',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Preço', fontsize=12)
plt.ylabel('Avaliação', fontsize=12)
plt.show()
```

## BOXPLOT

- boxplot = diagrama de caixa

O boxplot é um resumo visual. Mostra onde está a maioria dos dados

### paletas de cores prontas
- palette (propriedade)
- 'pastel' - Paleta: PASTEL (cores claras e suaves)
- 'deep' - Paleta: DEEP (cores vibrantes)
- 'Set2' - Paleta: Set2 cores suaves e pastéis (equilibrada, agradável)
- 'muted' - Paleta: MUTED (cores suaves e apagadas

```python
# Boxplot simples

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='periodo', y='vendas', palette='muted')
plt.title('Vendas por período',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Período', fontsize=12)
plt.ylabel('Vendas', fontsize=12)
plt.show()
```

## Histograma

- agrupa valores parecidos, mostra quantos tem de cada

```python
# Histograma
# ===========================

#bins = dividir os dados em grupos

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='avaliacao', bins=5)
plt.title('Distribuição das avaliações',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Avaliação', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.show()
```

```python
# Histograma com linha de densidade
# ===========================

# KDE = Kernel Density Estimation => estimativa de densidade

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='avaliacao', bins=5, kde=True)
plt.title('Distribuição das avaliações',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Avaliação', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.show()
```

```python
# Histograma com comparando períodos
# ===========================

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='vendas', hue='periodo', bins=5)
plt.title('Vendas: manhã vs tarde',
          fontsize=14,
          fontweight='bold')
plt.xlabel('Vendas', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.legend(title='Período')
plt.show()
```

#PARTE 3: Gráfico de Dispersão
1. Criar um gráfico de dispersão simples relacionando ticket_medio e satisfacao
2. Adicionar título, rótulos nos eixos X e Y
3. Adicionar cores por categoria usando o parâmetro hue='secao'
4. DESAFIO: Adicione tamanho às bolinhas baseado em vendas usando size='vendas'

Pergunta para refletir: Quando o ticket médio é maior, a satisfação aumenta ou diminui?

```python
# ============================
# PARTE 3: Gráfico de Dispersão
# ============================

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='ticket_medio',
    y='satisfacao',
    hue='secao',        # cores por seção (Feminino/Masculino) [file:135]
    size='vendas',      # DESAFIO: tamanho pela quantidade de vendas
    sizes=(20, 200),
    palette='Set2'
)
plt.title('Ticket médio x Satisfação por Seção')
plt.xlabel('Ticket médio (R$)')
plt.ylabel('Satisfação (0–10)')
plt.legend(title='Seção')
plt.tight_layout()
plt.show()
```

**Resposta:** No gráfico de dispersão, quando o ticket médio aumenta, a satisfação tende a **aumentar** também, pois os pontos com ticket médio mais alto aparecem, em geral, próximos dos valores mais altos de satisfação no eixo Y.

#PARTE 4: Boxplot
1. Criar um boxplot comparando vendas por secao (Feminino vs Masculino)

**Resposta:** Pelo boxplot, a seção **que vende mais** é aquela cujo boxplot está mais alto no eixo de vendas (maiores medianas e máximos); no seu gráfico, isso corresponde à seção que aparece com os valores centrais mais elevados. A seção com vendas mais **estáveis** é a que tem o box (caixa) e os “bigodes” mais **curtos**, indicando menor variação entre as vendas dos dias.

```python
# ==========================================
# PARTE 1: PREPARAR OS DADOS
# ==========================================

# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados da loja de roupas
df = pd.read_csv('loja_de_roupas.csv')

# Visualizando os dados
print("Dados da Loja de Roupas:")
print(df)
print("\n")

# ==========================================
# PARTE 2: ESCOLHER UM TEMA
# ==========================================

# Aplicando o tema whitegrid
sns.set_style('whitegrid')
print("Tema aplicado")

# ==========================================
# PARTE 3: GRÁFICO DE DISPERSÃO
# ==========================================

print("\n=== GRÁFICO DE DISPERSÃO ===\n")

# 1. Gráfico simples: ticket_medio x satisfacao
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ticket_medio', y='satisfacao')
plt.title('Ticket Médio x Satisfação dos Clientes', fontsize=14, fontweight='bold')
plt.xlabel('Ticket Médio (R$)')
plt.ylabel('Satisfação')
plt.show()

# 2. Com cores por seção
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ticket_medio', y='satisfacao', hue='secao')
plt.title('Ticket Médio x Satisfação por Seção', fontsize=14, fontweight='bold')
plt.xlabel('Ticket Médio (R$)')
plt.ylabel('Satisfação')
plt.legend(title='Seção')
plt.show()

# 3. DESAFIO: Com tamanhos baseados em vendas
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ticket_medio', y='satisfacao',
                hue='secao', size='vendas', sizes=(100, 500))
plt.title('Ticket Médio, Satisfação e Volume de Vendas', fontsize=14, fontweight='bold')
plt.xlabel('Ticket Médio (R$)')
plt.ylabel('Satisfação')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Gráficos de dispersão criados")
print("Observação: Parece que quando o ticket médio é maior, a satisfação também tende a ser maior")

# ==========================================
# PARTE 4: BOXPLOT
# ==========================================

print("\n=== BOXPLOT ===\n")

# 1. Boxplot simples: vendas por seção
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='secao', y='vendas')
plt.title('Distribuição de Vendas por Seção', fontsize=14, fontweight='bold')
plt.xlabel('Seção')
plt.ylabel('Vendas')
plt.show()

# 2. Com paleta de cores
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='secao', y='vendas', palette='Set2')
plt.title('Distribuição de Vendas por Seção', fontsize=14, fontweight='bold')
plt.xlabel('Seção')
plt.ylabel('Vendas')
plt.show()

print("Boxplots criados")
print("Observação: As duas seções têm vendas parecidas, mas a seção Feminina tem alguns dias com vendas muito altas")

# ==========================================
# PARTE 5: HISTOGRAMA
# ==========================================

print("\n=== HISTOGRAMA ===\n")

# 1. Histograma da satisfação
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='satisfacao', bins=5)
plt.title('Distribuição da Satisfação dos Clientes', fontsize=14, fontweight='bold')
plt.xlabel('Satisfação')
plt.ylabel('Quantidade')
plt.show()

# 2. Com linha KDE
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='satisfacao', bins=5, kde=True)
plt.title('Distribuição da Satisfação (com linha KDE)', fontsize=14, fontweight='bold')
plt.xlabel('Satisfação')
plt.ylabel('Quantidade')
plt.show()

# 3. Comparando vendas por seção
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='vendas', hue='secao', bins=5)
plt.title('Distribuição de Vendas: Feminino vs Masculino', fontsize=14, fontweight='bold')
plt.xlabel('Vendas')
plt.ylabel('Quantidade')
plt.legend(title='Seção')
plt.show()

print("Histogramas criados!")
print("Observação: A satisfação dos clientes está bem concentrada entre 4.0 e 4.8")
```
