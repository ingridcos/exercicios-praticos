```python
print("=== MANIPULANDO ARRAYS - ECOSSISTEMAS ===\n")

# Array espécies (ID, qtd, profundidade, tamanho) - reconstruído do PDF
especies = np.array([
    [747, 89, 33, 5], [623, 123, 32, 13], [501, 22, 49, 2], [116, 101, 42, 10],
    [297, 56, 69, 22], [613, 64, 27, 7], [295, 84, 29, 14], [692, 105, 72, 16],
    [229, 103, 35, 5], [374, 124, 70, 1]
])
print("Array espécies original:\n", especies, "\n")
```

```python
# 1a. Segunda coluna (quantidade)
qtd_especies = especies[:, 1]
print("1a. Quantidade de espécies:", qtd_especies, "\n")
```

```python
# 1b. Primeiras 3 quantidades
print("1b. Primeiras 3 quantidades:", qtd_especies[:3], "\n")
```

```python
# 1c. Últimas 5 quantidades
print("1c. Últimas 5 quantidades:", qtd_especies[-5:], "\n")
```

```python
# 1d. Tamanhos ordenados
tamanhos = especies[:, 3]
tamanhos_ordenados = np.sort(tamanhos)
print("1d. Tamanhos ordenados:", tamanhos_ordenados, "\n")
```

```python
# 2a. Index boolean - maior espécie (tamanho == 22)
maior_especie = especies[especies[:, 3] == 22]
print("2a. Maior espécie (tamanho 22):\n", maior_especie, "\n")
```

```python
# 2b. Fancy index - ID 297
id_297 = especies[especies[:, 0] == 297]
print("2b. Espécie ID 297:\n", id_297, "\n")
```

```python
# 2c. np.where - qtd 105
idx_105 = np.where(especies[:, 1] == 105)
linha_105 = especies[idx_105]
print("2c. Espécie com 105 representantes:\n", linha_105, "\n")
```

```python
# 2d. Profundidade > 60 = "Profundo"

# garante que especies existe antes deste trecho
# print(especies)
# print(especies.dtype)

# pega a coluna de profundidade e transforma em objeto (aceita int + str)
profundidade = especies[:, 2].astype(object)
profundidade[profundidade.astype(int) > 60] = "Profundo"

# copia o array original, mas como object para aceitar texto
especies_copy = especies.astype(object)
especies_copy[:, 2] = profundidade

print("2d. Profundidade (>60 = Profundo):\n", especies_copy[:, 2], "\n")
```

```python
# 3a. Adicionar 2 novas espécies
novas_especies = np.array([[204, 10, 40, 12], [392, 11, 81, 11]])
especies_expandida = np.vstack([especies, novas_especies])
print("3a. +2 espécies (primeiras 5):\n", especies_expandida[:5], "\n")
```

```python
# 3b. Nova coluna visão [0,1,0,0,0,0,1,0,1,1,0]
# essa lista tem que ter o MESMO número de linhas do array
visao = np.array([0, 1, 0, 0, 0, 0, 1, 0, 1, 1])  # para 10 linhas
# se você tiver 11 linhas, use:
# visao = np.array([0, 1, 0, 0, 0, 0, 1, 0, 1, 1, 0])

# empilha a nova coluna no array com profundidade tratada
especies_com_visao = np.column_stack([especies_copy, visao])

print("3b. + Coluna visão (primeiras 3 linhas):\n", especies_com_visao[:3], "\n")
```

```python
# Array acidentes
acidentes = np.array([1, 3, 2, 0, 1, 0, 2, 1, 4, 0, 0, 0, 1, 1, 0])
print("Acidentes:", acidentes, "\n")
```

```python
# 1. Clientes abaixo da média
media_acidentes = np.mean(acidentes)
desconto = acidentes[acidentes < media_acidentes]
print("1. Abaixo da média (", round(media_acidentes, 2), "):", desconto, "\n")
```

```python
#2. índices (clientes) que tiveram pelo menos 2 anos sem acidentes
clientes_2_anos_sem = np.where(acidentes == 0)[0]  # aqui são todos que tiveram 0 pelo menos 1 vez

print("2. Clientes com pelo menos 1 ano sem acidentes (zeros):", clientes_2_anos_sem, "\n")
```

```python
# 3. Função 3x - 2y + xy (exemplo arrays)
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
resultado = 3*x - 2*y + x*y
print("3. 3x-2y+xy:", resultado, "\n")
```

```python
# 4. Adicionar notas grupo [1, 2, 1] às provas
# cada linha = 1 estudante, cada coluna = 1 prova
notas_provas = np.array([
    [5, 6, 7],
    [8, 9, 10],
    [7, 8, 9]
])

# mesma nota de trabalho por prova: 1, 2, 1
notas_grupo = np.array([1, 2, 1])

print("Shape notas_provas:", notas_provas.shape)  # (3, 3)
print("Shape notas_grupo:", notas_grupo.shape)    # (3,)

# broadcasting: soma o vetor [1,2,1] em cada linha
notas_finais = notas_provas + notas_grupo

print("4. Notas finais + grupo:\n", notas_finais, "\n")
```
