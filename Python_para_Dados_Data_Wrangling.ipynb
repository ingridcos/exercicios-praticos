```python
import pandas as pd

# Re-load df and apply previous drop operation to ensure df is not None
df = pd.read_csv("/content/collected research people.csv")
df = df.drop(columns=['Unnamed: 0'])

novas_cols = {'Endereerço': 'Endereço',
              'phone': 'Telefone',
              'anual_income': 'Renda Anual',
              'AAge': 'Idade',
              'n_filhos': 'N Filhos',
              'diariamente_sun blocker': 'Usa protetor solar diariamente',
              'Food_restrictions': 'Restrição alimentar',
              'streaming': 'Streaming'}

df = df.rename(columns=novas_cols)
df.head()
```

```python
df.Nome.str.replace('\W', '', regex=True)
```

```python
df['Nome'] = df['Nome'].str.replace('[^a-zA-Zà-úÀ-Ú0-9 ]', '', regex=True)
```

```python
df.Nome = df.Nome.str.replace('[^a-zA-Zà-úÀ-Ú0-9 ', ' ')
df.head()
```

```python
df.Telefone.str.replace('[^0-9()]', ' ')
```

```python
import re

def limpa_telefone(valor):
    if pd.isna(valor):
        return pd.NA

    # deixa só dígitos
    nums = re.sub(r'\D', '', str(valor))

    # remove 55 extra no início, se tiver
    if nums.startswith('55') and len(nums) > 11:
        nums = nums[2:]

    # se tiver 10 dígitos (fixo), força um 9 na frente
    if len(nums) == 10:
        ddd = nums[:2]
        numero = '9' + nums[2:]
    # se tiver 11 dígitos (celular)
    elif len(nums) == 11:
        ddd = nums[:2]
        numero = nums[2:]
    else:
        return pd.NA  # joga fora casos muito zoados

    # separa 9 + 4 + 4
    return f'+55 ({ddd}) {numero[0]}{numero[1:5]}-{numero[5:]}'
```

```python
df['Telefone'] = df['Telefone'].apply(limpa_telefone)
df.head()
```
