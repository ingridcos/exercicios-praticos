**Outliers:**
*   Limite Superior e Inferior
*   Quantidade de Outliers

Preços ordenados:

$1k → $5k (25%)

$6.9k→ $12k → $18.2k (50%)

$35k → $100k (75%)

1. Q1 = 25% dos dados (precos.quantile(0.25))
2. Q3 = 75% dos dados (precos.quantile(0.75))  
3. IQR = Q3 - Q1 (intervalo do meio)
4. LIM_INFERIOR = Q1 - 1.5 × IQR
5. LIM_SUPERIOR = Q3 + 1.5 × IQR
6. OUTLIERS = valores < LIM_INFERIOR OU > LIM_SUPERIOR

```python
precos = df_bronze["preco_venda"]
Q1 = precos.quantile(0.25)
Q3 = precos.quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR

print(f"Q1 real: ${Q1:,.0f}")
print(f"Q3 real: ${Q3:,.0f}")
print(f"IQR real: ${IQR:,.0f}")
print(f"Limite superior CORRETO: ${limite_superior:,.0f}")

outliers = df_bronze[df_bronze["preco_venda"] > limite_superior]
print(f"Outliers reais: {len(outliers)}")
print(f"Percentual: {len(outliers)/len(df_bronze)*100:.1f}%")
print(f"Total linhas: {len(df_bronze)}")
```
