## Conceitos de FIGURE e AXES

- Figure -> Figura = A folha de papel completa
- Axes (eixos) -> O gráfico desenha dentro dessa folha, área onde os dados aparecem

```python
fig = plt.figure(figsize=(10,6), facecolor='lightgray')

# ax = a ÁREA do gráfico dentro da folha
ax = fig.add_subplot(111)

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Usamos ax.plot() ao invés de plt.plot()
ax.plot(x, y,
        linewidth=3,
        marker='o')

ax.set_title('Este é o AXES (área do gráfico)')

ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')

# plt.suptitle() = super título
plt.suptitle('A área cinza é a FIGURE (folha completa)',
             fontsize=14,
             fontweight='bold')

plt.show()
```

## QUANDO USAR CADA UM?

### plt.plot() - AUTOMÁTICO
- Gráficos simples e rápidos
- Quando você vai fazer só 1 gráfico
- Para praticar e aprender
- Quando o padrão já está bom

### fig, ax - MANUAL  
- Quando quiser controlar o tamanho (figsize)
- Quando for fazer VÁRIOS gráficos juntos (subplots!)
- Quando precisar de ajustes finos
- Códigos mais profissionais

### DICA: Para subplots, você PRECISA usar o jeito manual

- Subplots => mais de 1 gráficos, vários gráficos em uma mesma figura


```python
# EXEMPLO 1 - 2 GRÁFICOS LADO A LADO
# ====================================

# dados de vendas de duas lojas
dias = ['seg', 'ter', 'qua', 'qui', 'sex']
loja_1= [150, 200, 180, 220, 250]
loja_2= [180, 170, 200, 240, 280]

# Criar figure com 2 subplots (1 linha e 2 colunas)
# axes agora é uma LISTA / VETOR com 2 axes [ax1, ax2]
fig, axes = plt.subplots(1, 2, figsize=(14,5))


# ==============================================
# Grafico 1 (loja 1) - primeira posição
# ==============================================

axes[0].plot(dias, loja_1,
             color='purple',
             linewidth=2,
             marker='o',
             markersize=8)

axes[0].set_title('Vendas da loja 1',
                  fontsize=14,
                  fontweight='bold')


axes[0].set_xlabel('Dia da semana', fontsize=12)

axes[0].set_ylabel('Vendas (R$)', fontsize=12)

axes[0].grid(True, alpha=0.3)



# ==============================================
# Grafico 2 (loja 2) - segunda posição
# ==============================================

axes[1].plot(dias, loja_2,
             color='green',
             linewidth=2,
             marker='s',
             markersize=8)

axes[1].set_title('Vendas da loja 2',
                  fontsize=14,
                  fontweight='bold')


axes[1].set_xlabel('Dia da semana', fontsize=12)

axes[1].set_ylabel('Vendas (R$)', fontsize=12)


axes[1].grid(True, alpha=0.3)


fig.suptitle('Comparação de vendas entre lojas',
             fontsize=16,
             fontweight='bold')

# Organizador automático - harmoniza os espaços automaticamente
plt.tight_layout()
plt.savefig('vendas_semanais.png',
            dpi=300,
            bbox_inches='tight',
            facecolor='white')
plt.show()
```

## DETALHES IMPORTANTES

- subplots(1, 2) = 1 linha, 2 colunas
- axes[0] = 1º gráfico
- axes[1] = 2º gráfico
- fig.suptitle() = título GERAL da figura inteira"
- Salvar como PNG de qualidade: `plt.savefig()`
- dpi = dots per inch (pontos por polegada)
  <br> 72-100 = boa para web
  <br>150-200 = boa para apresentações
  <br>300+ = alta qualidade para impressão
- bbox_inches='tight' = corta espaços brancos desnecessários
- facecolor='white' = fundo da imagem (se não colocar, pode ficar transparente)"

##Tarefas:
#PARTE 1: Preparar os Dados
1. Importar matplotlib e matplotlib.cm
2. Criar lista com dias da semana (Segunda, Terça, … ,Domingo)
3. Criar três listas com horas de estudo por dia. Ex.:
● Python (ex: 2, 3, 2, 4, 3, 5, 4)
● SQL (ex: 1, 2, 3, 2, 3, 4, 3)
● Estatística (ex: 1, 1, 2, 2, 2, 3, 2)


```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Lista com dias da semana
dias_semana = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']

# Listas com horas de estudo por dia
horas_python =     [2, 3, 2, 4, 3, 5, 4]
horas_sql =        [1, 2, 3, 2, 3, 4, 3]
horas_estatistica = [1, 1, 2, 2, 2, 3, 2]
```

#PARTE 2: Criar Subplots
1. Criar uma figure com 2 subplots lado a lado (1 linha, 2 colunas)
2. Definir figsize apropriado (ex: 16x6)
3. Escolher uma paleta de cores (viridis ou plasma) com 3 cores

```python
# PARTE 2: Criar Subplots

# 1. Figure com 2 subplots lado a lado (1 linha, 2 colunas) e figsize 16x6
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 2. Escolher paleta de cores (ex: viridis) com 3 cores
cores = cm.viridis([0.2, 0.5, 0.8])
```

#PARTE 3: Primeiro Gráfico - Linhas Comparativas

1. No primeiro subplot (axes[0]):

● Plotar as 3 linhas (Python, SQL, Estatística)

● Usar as cores da paleta escolhida

● Adicionar marcadores diferentes para cada linha (o, s, ^)

● Adicionar título: "Horas de Estudo por Dia"

● Adicionar rótulos nos eixos (X: "Dia da Semana", Y: "Horas")

● Adicionar legenda

● Adicionar grid

```python
# PARTE 3: Primeiro Gráfico - Linhas Comparativas

ax1 = axes[0]  # primeiro subplot

# Plotar as 3 linhas com cores da paleta e marcadores diferentes
ax1.plot(dias_semana, horas_python,
         color=cores[0], marker='o', label='Python')
ax1.plot(dias_semana, horas_sql,
         color=cores[1], marker='s', label='SQL')
ax1.plot(dias_semana, horas_estatistica,
         color=cores[2], marker='^', label='Estatística')

# Título e rótulos dos eixos
ax1.set_title('Horas de Estudo por Dia')
ax1.set_xlabel('Dia da Semana')
ax1.set_ylabel('Horas')

# Legenda e grid
ax1.legend()
ax1.grid(True, alpha=0.3)
```

#PARTE 4: Primeiro Gráfico - Linhas Comparativas
1. No segundo subplot (axes[1]):

● Calcular o total de horas de cada matéria (use sum())

● Criar gráfico de barras com os 3 totais

● Usar as mesmas cores da paleta

● Adicionar título: "Total de Horas na Semana"

● Adicionar rótulo no eixo Y: "Total de Horas"

● Adicionar grid no eixo Y

● EXTRA: Colocar os valores em cima de cada barra


```python
# PARTE 4: Segundo Gráfico - Barras com Totais

ax2 = axes[1]  # segundo subplot

# Calcular o total de horas de cada matéria
total_python = sum(horas_python)
total_sql = sum(horas_sql)
total_estatistica = sum(horas_estatistica)

materias = ['Python', 'SQL', 'Estatística']
totais = [total_python, total_sql, total_estatistica]

# Gráfico de barras com as mesmas cores da paleta
barras = ax2.bar(materias, totais, color=cores)

# Título e rótulo do eixo Y
ax2.set_title('Total de Horas na Semana')
ax2.set_ylabel('Total de Horas')

# Grid apenas no eixo Y
ax2.grid(axis='y', alpha=0.3)

# EXTRA: Colocar os valores em cima de cada barra
for barra in barras:
    altura = barra.get_height()
    ax2.text(
        barra.get_x() + barra.get_width() / 2,
        altura + 0.1,
        f'{altura}',
        ha='center',
        va='bottom'
    )
```

#PARTE 5: Finalizar
1. Adicionar título geral na figure: "Meu Dashboard de Estudos - Semana 1"
2. Aplicar plt.tight_layout()
3. Salvar como 'dashboard_estudos.png' com dpi=300
4. Exibir o gráfico final com plt.show()


```python
# PARTE 5: Finalizar

fig.suptitle('Meu Dashboard de Estudos - Semana 1', fontsize=16)

plt.tight_layout()

plt.savefig('dashboard_estudos.png', dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # paletas de cores

# ============================
# PARTE 1: Preparar os dados
# ============================

dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']

horas_python =     [2, 3, 2, 4, 3, 5, 4]
horas_sql =        [1, 2, 3, 2, 3, 4, 3]
horas_estatistica = [1, 1, 2, 2, 2, 3, 2]

# Paleta de cores (viridis com 3 cores) [file:125]
cores = cm.viridis([0.2, 0.5, 0.8])

# Totais por matéria (para o gráfico de barras)
total_python = sum(horas_python)
total_sql = sum(horas_sql)
total_estatistica = sum(horas_estatistica)

materias = ['Python', 'SQL', 'Estatística']
totais = [total_python, total_sql, total_estatistica]

# ============================
# PARTE 2: Criar subplots
# ============================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 linha, 2 colunas [file:125]

# ============================
# PARTE 3: Gráfico de linhas
# ============================

ax1 = axes[0]

ax1.plot(dias, horas_python,
         color=cores[0],
         marker='o',
         label='Python')

ax1.plot(dias, horas_sql,
         color=cores[1],
         marker='s',
         label='SQL')

ax1.plot(dias, horas_estatistica,
         color=cores[2],
         marker='^',
         label='Estatística')

ax1.set_title('Horas de Estudo por Dia')
ax1.set_xlabel('Dia da Semana')
ax1.set_ylabel('Horas')
ax1.grid(True, alpha=0.3)
ax1.legend()

# ============================
# PARTE 4: Gráfico de barras
# ============================

ax2 = axes[1]

barras = ax2.bar(materias, totais, color=cores)

ax2.set_title('Total de Horas na Semana')
ax2.set_ylabel('Total de Horas')
ax2.grid(axis='y', alpha=0.3)

# EXTRA: colocar valores em cima de cada barra
for barra in barras:
    altura = barra.get_height()
    ax2.text(barra.get_x() + barra.get_width()/2,
             altura + 0.1,
             f'{altura}',
             ha='center', va='bottom')

# ============================
# PARTE 5: Finalizar
# ============================

fig.suptitle('Meu Dashboard de Estudos - Semana 1', fontsize=16)  # título geral [file:125]
plt.tight_layout()
plt.savefig('dashboard_estudos.png', dpi=300)  # salva em alta qualidade [file:125]
plt.show()
```

```python
# 1. IMPORTAR
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 2. PREPARAR DADOS
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex']
valores1 = [10, 20, 15, 25, 30]
valores2 = [15, 18, 20, 22, 28]

# 3. CRIAR SUBPLOTS
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 4. ESCOLHER CORES
cores = cm.viridis([0.3, 0.7])

# 5. FAZER GRÁFICOS
axes[0].plot(dias, valores1, color=cores[0])
axes[0].set_title('Gráfico 1')
axes[0].set_xlabel('Dias')
axes[0].set_ylabel('Valores')
axes[0].grid(True, alpha=0.3)

axes[1].bar(dias, valores2, color=cores[1])
axes[1].set_title('Gráfico 2')
axes[1].set_xlabel('Dias')
axes[1].set_ylabel('Valores')
axes[1].grid(True, alpha=0.3)

# 6. TÍTULO GERAL
fig.suptitle('Meu Dashboard', fontsize=16, fontweight='bold')

# 7. ORGANIZAR
plt.tight_layout()

# 8. SALVAR
plt.savefig('dashboard.png', dpi=300, bbox_inches='tight')

# 9. MOSTRAR
plt.show()
```
