## Descobrindo o Dia da Semana üóìÔ∏è

Python consegue nos dizer que dia da semana √© hoje, usando `strftime()`

Os c√≥digos s√£o:
- `%A` = Nome completo do dia (segunda-feira)
- `%a` = Nome abreviado do dia (seg, ter, qua...)
- `%d` = Dia do m√™s (01, 02, ... 15)
- `%m` = M√™s (01, 02...12)
- `%y` = Ano com 4 d√≠gitos (2025)

```python
dia_semana = agora.strftime("%A")
dia_semana_abreviado = agora.strftime("%a")
dia = agora.strftime("%d")
mes = agora.strftime("%m")
ano = agora.strftime("%y")

print(f"Hoje √© dia: {dia_semana}")
print(f"Hoje √© dia abreviado: {dia_semana_abreviado}")

#\n para quebrar linha
print(f"\nHoje √© : {dia}/{mes}/{ano}")
```

## Calculando Idade em Dias üéÇ

Agora vamos calcular quantos dias eu j√° vivi!

Para isso, vamos
1. Criar uma vari√°vel com a data de nascimento
2. Subtrair da data atual
3. Ver o resultado em dias

```python
# Minha data de nascimento 16/07/1990
# Formato date(ano, mes, dia)

data_nasc = date(1997, 7, 16)

# Data de hoje
hoje = date.today()

# Calculando a diferen√ßa
diferenca = hoje - data_nasc

# Quantos dias eu vivi?
dias_vividos = diferenca.days

print(f"Data de nascimento: {data_nasc.strftime('%d/%m/%Y')}")
print(f"Data de hoje: {hoje.strftime('%d/%m/%Y')}")
print(f"\nEu j√° vivi: {dias_vividos} dias!")
print(f"\nIsso equivale a aproximadamente: {dias_vividos // 365} anos!")
```

## Finalizando

Ap√≥s calcular a quantidade de dias vividos, podemos dividir o resultado dias_vividos / 365 para descobrir quantos anos j√° vivemos!

#Relat√≥rio


```python
from datetime import date

nome = "Ingrid"
cidade = "S√£o Paulo"

data_nasc = date(1997, 7, 16)
hoje = date.today()

diferenca = hoje - data_nasc
dias_vividos = diferenca.days
anos_vividos = dias_vividos // 365

relatorio = f"""
========== RELAT√ìRIO DE VIDA ==========
Nome: {nome}
Cidade: {cidade}
Data de nascimento: {data_nasc.strftime('%d/%m/%Y')}
Data de hoje: {hoje.strftime('%d/%m/%Y')}

Voc√™ j√° viveu: {dias_vividos} dias.
Isso equivale a aproximadamente: {anos_vividos} anos.
=======================================
"""

print(relatorio)
```

```python
# ====================================================
# NOSSO PRIMEIRO GR√ÅFICO
# ====================================================

import matplotlib.pyplot as plt

print("Biblioteca importada com sucesso! ‚úÖ\n")

# Contexto: Entender como as horas de sono afetam o n√≠vel de energia

print("-" * 55)
print("EXEMPLO 1: Gr√°fico simples - Horas de sono x Energia")
print("-" * 55)

# Dados da semana (cada posi√ß√£o = um dia)

horas_sono = [4, 5, 6, 7, 8, 6, 5]
nivel_energia = [3, 4, 6, 8, 9, 6, 4]
print("\n")  # quebra de linha


plt.title('Rela√ß√£o entre Horas de Sono e N√≠vel de Energia', fontsize=14, fontweight='bold')

# fontsize = tamanho da fonte
# fontweight = peso da fonte

# R√≥tulo no eixo X
plt.xlabel('Horas de Sono', fontsize=12)

# R√≥tulo no eixo Y
plt.ylabel('N√≠vel de Energia (0-10)', fontsize=12)

# Grade de fundo
# alpha deixa a grade mais transparente

plt.grid(True, alpha=0.3)

# Criando o gr√°fico
plt.plot(horas_sono, nivel_energia)

# Exibir o gr√°fico na tela
plt.show()
```

## Adicionar T√≠tulos e R√≥tulos

- Para adicionar t√≠tulo: `plt.title('texto do t√≠tulo')`
- Para adicionar r√≥tulo no eixo x: `plt.xlabel('texto do r√≥tulo')`
- Para adicionar r√≥tulo no eixo y: `plt.ylabel('texto do r√≥tulo')`
- Para adicionar grade: `plt.grid(True)`


## Personalizando Cores e Estilo

- color = cor  
- espessura / largura da linha = linewidth (width = largura)  
- estilo da linha = linestyle  
- marcadores de pontos (bolinhas) = marker  
- tamanho do marcador = markersize  
- cor dos marcadores = markerfacecolor (cor de preenchimento dos pontos)  
- legenda = label  


```python
# Contexto: Entender como as horas de sono afetam o n√≠vel de energia

horas_sono = [4, 5, 6, 7, 8, 6, 5]
nivel_energia = [3, 4, 6, 8, 9, 6, 4]

plt.title(
    'Rela√ß√£o entre Horas de Sono e N√≠vel de Energia',
    fontsize=14,
    fontweight='bold',
    color='darkblue'
)


# fontsize = tamanho da fonte
# fontweight = peso da fonte

# R√≥tulo no eixo X
plt.xlabel(
    'Horas de Sono',
    fontsize=12,
    color='blue'
)


# R√≥tulo no eixo Y
plt.ylabel(
    'N√≠vel de Energia (0-10)',
    fontsize=12,
    color='blue'
)


# Grade de fundo
# alpha deixa a grade mais transparente
plt.grid(True, alpha=0.3)

# Criando o gr√°fico
plt.plot(horas_sono, nivel_energia,
         color='purple',
         linewidth=3,
         linestyle='--',
         marker='o',
         markersize=8,
         markerfacecolor='pink',
         label='Minha Semana')

# plotar = desenhar o gr√°fico

# Exibindo o gr√°fico na tela
plt.show()
```

## Figure e Axes

- Figure = Figura => a folha de papel inteira
- Axes = Eixos => O gr√°fico/desenho dentro dessa folha, a √°rea onde os dados aparecem

diferen√ßa entre usar plt e ax
- plt √© o jeito autom√°tico de criar o gr√°fico => gr√°ficos simples, s√≥ 1 gr√°fico
- ax √© o manual = maior controle do gr√°fico => quer controlar tamanho da figura (figsize), gr√°ficos mais complexos, v√°rios gr√°ficos juntos (subplots)

```python
import matplotlib.pyplot as plt

horas_sono = [4, 5, 6, 7, 8, 6, 5]
nivel_energia = [3, 4, 6, 8, 9, 6, 4]

# Modo simples com plt
plt.plot(horas_sono, nivel_energia)
plt.title('Horas de Sono x N√≠vel de Energia')
plt.xlabel('Horas de Sono')
plt.ylabel('N√≠vel de Energia (0-10)')
plt.grid(True, alpha=0.3)
plt.show()

# Modo com Figure e Axes (ax)
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(
    horas_sono, nivel_energia,
    color='purple',
    marker='o',
    linestyle='--'
)
ax.set_title('Horas de Sono x N√≠vel de Energia')
ax.set_xlabel('Horas de Sono')
ax.set_ylabel('N√≠vel de Energia (0-10)')
ax.grid(True, alpha=0.3)

plt.show()
```

```python
horas_sono =    [4, 5, 6, 7, 8, 6, 5]
nivel_energia = [3, 4, 6, 8, 9, 6, 4]

# Criando figure e axes explicitamente
# fig = moldura
# ax = a √°rea do gr√°fico dentro da folha
# figsize = tamanho da folha em polegadas (largura, altura)

fig, ax = plt.subplots(figsize=(10,6))

# Usar ax.plot() ao inv√©s de plt.plot()
ax.plot(horas_sono, nivel_energia,
        color='purple',
        linewidth=3,
        marker='D',
        markersize=8)

# plt.title() vira ax.set_title()
ax.set_title('Gr√°fico usando axes e figure - Horas de sono',
             fontsize=14,
             fontweight='bold')

# plt.xlabe() vira ax.set_xlabel()
ax.set_xlabel('Horas de sono',
             fontsize=12,
             fontweight='bold')

# plt.ylabe() vira ax.set_ylabel()
ax.set_ylabel('N√≠vel de energia',
             fontsize=12,
             fontweight='bold')

# plt.grid() vira ax.grid()
```
