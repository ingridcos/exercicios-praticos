```python
# importando bibliotecas

import pandas as pd
import plotly.express as px

print("Bibliotecas importadas! ✅")

df = pd.read_csv("academia.csv")
print(df)
```

```python
# Gráfico de Dispersão - Scatter
# quando tem mais aulas, vem mais alunos?

fig = px.scatter(
                  df,
                  x="aulas",
                  y="alunos",
                  title="Relação entre Aulas e Alunos",
                )
fig.show()
```

```python
# adicionar cores para separar os períodos (manhã e tarde)

# hue do seaborn = color => semelhantes
# scatter = dispersão

fig = px.scatter(
                  df,
                  x="aulas",
                  y="alunos",
                  title="Relação entre Aulas e Alunosr",
                  color='periodo',
                  labels={
                      'aulas': 'Quantidade de aulas',
                      'alunos': 'Quantidade de alunos',
                      'periodo': 'Período'
                  }
                )
fig.show()
```

```python
# Gráfico de linhas
# movimento durante a semana
# line = linha

fig = px.line(
                  df,
                  x="dia",
                  y="alunos",
                  color='periodo',
                  title="Alunos por periodo durante a semana",
                  labels={
                      'dia': 'Dia da semana',
                      'alunos': 'Quantidade de alunos',
                      'periodo': 'Período'
                  },
                  markers=True
                )
fig.show()
```

```python
# Gráfico de barras
# Comparar dias
# bar = barras
# barmode='group' => modo / estilo da barra = agrupado, lado a lado

fig = px.bar(
                  df,
                  x="dia",
                  y="alunos",
                  color='periodo',
                  title="Alunos por Dia e Período",
                  labels={
                      'dia': 'Dia da semana',
                      'alunos': 'Quantidade de alunos',
                      'periodo': 'Período'
                  },

                  barmode='group' # lado a lado, usa para comparar categorias
                )
fig.show()
```

```python
# Gráfico de barras

fig = px.bar(
                  df,
                  x="dia",
                  y="alunos",
                  color='periodo',
                  title="Alunos por Dia e Período",
                  labels={
                      'dia': 'Dia da semana',
                      'alunos': 'Quantidade de alunos',
                      'periodo': 'Período'
                  },
                  barmode='stack' # empilhado, usa para ver o total

                )
fig.show()
```

```python
fig.write_html("dashboard_academia.html")
print("Gráfico salvo como html com sucesso! ✅")
```

Dica: O arquivo tem dados de vendas de uma semana com: dia, vendas (R$), produtos
(quantidade), satisfacao (0-5) e categoria (Olhos/Lábios/Rosto)

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('loja_maquiagem.csv')
print(df)
```

#PARTE 2: Gráfico de Dispersão
1. Criar scatter plot relacionando vendas (x) e satisfacao (y)
2. Adicionar cores por categoria usando color="categoria"
3. Adicionar tamanho das bolinhas baseado em produtos usando size="produtos"
4. Adicionar título e rótulos bonitos com labels
5. Adicionar hover_data=["dia"] para mostrar o dia ao passar o mouse

Pergunta para refletir: Quando as vendas são maiores, a satisfação aumenta? Qual
categoria tem melhor relação vendas × satisfação?

```python
# ============================
# PARTE 2: Gráfico de Dispersão
# ============================

fig_scatter = px.scatter(
    df,
    x='vendas',
    y='satisfacao',
    color='categoria',
    size='produtos',
    hover_data=['dia'],
    labels={
        'vendas': 'Vendas (R$)',
        'satisfacao': 'Satisfação (0–5)',
        'categoria': 'Categoria',
        'produtos': 'Qtd. de Produtos',
        'dia': 'Dia da Semana'
    },
    title='Vendas x Satisfação por Categoria'
)
fig_scatter.show()
```

Resposta: Pelo gráfico de dispersão, dá para ver que, quando as vendas são maiores, a satisfação também tende a aumentar, porque os pontos com valores mais altos de vendas aparecem, em geral, junto de valores de satisfação acima de 4.

#PARTE 3: Gráfico de Linhas
1. Criar line plot com dia (x) e vendas (y)
2. Separar por categoria usando color="categoria"
3. Adicionar markers=True para ter bolinhas nos pontos
4. Personalizar labels para ficarem mais bonitos

Pergunta para refletir: Qual categoria vende mais durante a semana? Isso muda de
segunda para domingo?

```python
# PARTE 3: Gráfico de Linhas
# ============================

fig_line = px.line(
    df,
    x='dia',
    y='vendas',
    color='categoria',
    markers=True,
    labels={
        'dia': 'Dia da Semana',
        'vendas': 'Vendas (R$)',
        'categoria': 'Categoria'
    },
    title='Vendas por Dia e Categoria'
)
fig_line.show()
```

Resposta: A categoria que mais vende durante a semana é Lábios, porque a linha dessa categoria fica, na maior parte dos dias, acima das demais no eixo de vendas. De segunda a domingo isso não muda muito: Lábios continua liderando, mas Olhos e Rosto também crescem bastante no fim de semana, deixando as linhas mais próximas, principalmente no sábado e no domingo.

#PARTE 4: Gráfico de Barras
5. Criar bar plot comparando produtos (y) por dia (x)
6. Colorir por categoria
7. Usar barmode="group" para barras lado a lado
8. Adicionar título e rótulos

Pergunta para refletir: Qual dia vende mais produtos no total? Qual categoria domina cada
dia?

```python
# ============================
# PARTE 4: Gráfico de Barras
# ============================

fig_bar = px.bar(
    df,
    x='dia',
    y='produtos',
    color='categoria',
    barmode='group',
    labels={
        'dia': 'Dia da Semana',
        'produtos': 'Quantidade de Produtos',
        'categoria': 'Categoria'
    },
    title='Produtos Vendidos por Dia e Categoria'
)
fig_bar.show()
```

Resposta: O dia com maior quantidade total de produtos vendidos é o **sábado**, seguido de perto pelo **domingo**, pois as barras de todos os produtos nesse dia são as mais altas.
Em praticamente todos os dias, a categoria **Lábios** costuma ter a barra mais alta (mais produtos vendidos), com **Olhos** em segundo lugar e **Rosto** logo atrás, principalmente no fim de semana.

```python
# SOLUÇÃO DO EXERCÍCIO: LOJA DE MAQUIAGEM

# ===== PARTE 1: PREPARAR OS DADOS =====

import pandas as pd
import plotly.express as px

df = pd.read_csv('loja_maquiagem.csv')
print(df)


# ===== PARTE 2: GRÁFICO DE DISPERSÃO =====

fig = px.scatter(
    df,
    x='vendas',
    y='satisfacao',
    color='categoria',
    size='produtos',
    title='Relação entre Vendas e Satisfação por Categoria',
    labels={
        'vendas': 'Vendas (R$)',
        'satisfacao': 'Satisfação dos Clientes',
        'categoria': 'Categoria',
        'produtos': 'Produtos Vendidos'
    },
    hover_data=['dia']
)

fig.show()


# ===== PARTE 3: GRÁFICO DE LINHAS =====

fig = px.line(
    df,
    x='dia',
    y='vendas',
    color='categoria',
    title='Evolução das Vendas Durante a Semana',
    labels={
        'dia': 'Dia da Semana',
        'vendas': 'Vendas (R$)',
        'categoria': 'Categoria'
    },
    markers=True
)

fig.show()


# ===== PARTE 4: GRÁFICO DE BARRAS =====

fig = px.bar(
    df,
    x='dia',
    y='produtos',
    color='categoria',
    title='Produtos Vendidos por Dia',
    labels={
        'dia': 'Dia da Semana',
        'produtos': 'Quantidade de Produtos',
        'categoria': 'Categoria'
    },
    barmode='group'
)

fig.show()


# ===== PARTE 5: PERSONALIZAÇÃO =====

fig = px.scatter(
    df,
    x='vendas',
    y='satisfacao',
    color='categoria',
    size='produtos',
    title='Dashboard - Vendas da Loja de Maquiagem',
    labels={
        'vendas': 'Vendas (R$)',
        'satisfacao': 'Satisfação dos Clientes',
        'categoria': 'Categoria',
        'produtos': 'Produtos Vendidos'
    },
    hover_data=['dia']
)

fig.update_layout(
    font=dict(size=14),
    plot_bgcolor='#f8f9fa',
    paper_bgcolor='white',
    title_font_size=20,
    title_x=0.5,
    height=600
)

fig.update_xaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='lightgray'
)

fig.update_yaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='lightgray'
)

fig.show()


# ===== PARTE 6: EXPORTAR =====

fig.write_html('dashboard_maquiagem.html')

print("\nGráfico salvo como 'dashboard_maquiagem.html'!✅ ")
```
