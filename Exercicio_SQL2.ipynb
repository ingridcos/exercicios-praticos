```python
import sqlite3

conn = sqlite3.connect('bi_class.db')
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tabelas = cursor.fetchall()

for t in tabelas:
    print(t[0])
```

```python
import pandas as pd

for t in tabelas:
    nome = t[0]
    print(f'===== {nome} =====')
    df = pd.read_sql_query(f"SELECT * FROM {nome}", conn)
    print(df.head())   # ou print(df) se quiser tudo
```
