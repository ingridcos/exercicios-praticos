```python
import pandas as pd

clientes = pd.read_csv('CLIENTES.csv')
pedidos  = pd.read_csv('PEDIDOS.csv')
produtos = pd.read_csv('PRODUTOS.csv')

# 1) junta PEDIDOS com CLIENTES pelo ID_CLIENTE
pedidos_clientes = pedidos.merge(clientes, on='ID_CLIENTE', how='inner')

# 2) junta o resultado com PRODUTOS pelo ID_PRODUTO
tudo = pedidos_clientes.merge(produtos, on='ID_PRODUTO', how='inner')

tudo.head()
```

```python
tudo = tudo.rename(columns={'NOME_x': 'NOME_CLIENTE', 'NOME_y': 'NOME_PRODUTO'})

tudo = tudo[[
    'ID_CLIENTE', 'NOME_CLIENTE', 'EMAIL',          # colunas de CLIENTES
    'ID_PEDIDO', 'DATA_PEDIDO',             # colunas de PEDIDOS
    'ID_PRODUTO', 'NOME_PRODUTO', 'PRECO'   # colunas de PRODUTOS
]]

tudo.head()
```
